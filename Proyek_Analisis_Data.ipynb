### Cleaning Data

#### Day DF

```python
day_df['yr'] = day_df ['yr'].replace({0 : 2011, 1 : 2012})
```

```python
day_df['season'] = day_df['season'].replace({1 : "springer", 2 : "summer", 3 : "fall", 4 : "winter"})
```

```python
day_df['weathersit'] = day_df['weathersit'].replace({1 : "Clear", 2 : "Mist + Cloudy", 3 : "Light Snow", 4 : "Heavy Rain + Ice Pallets"})
```

```python
day_df['holiday'] = day_df['holiday'].replace({0 : "No", 1 : "Yes"})
```

```python
day_df['workingday'] = day_df['workingday'].replace({0 : "No", 1 : "Yes"})
```

```python
day_df['weekday'] = day_df['weekday'].replace({0 : "Sunday", 1 : "Monday", 2 : "Tuesday", 3 : "Wednesday", 4 : "Thursday", 5 : "Friday", 6 : "Saturday"})
```

```python
day_df['mnth'] = day_df['mnth'].replace({1 : "January", 2 : "February", 3 : "March", 4 : "April", 5 : "May", 6 : "June", 7 : "July", 8 : "August", 9 : "September",
                                         10 : "October", 11 : "November", 12 : "December"})
```

```python
day_df.head()
```

#### Hour DF

```python
hour_df['yr'] = hour_df ['yr'].replace({0 : 2011, 1 : 2012})
hour_df['season'] = hour_df['season'].replace({1 : "springer", 2 : "summer", 3 : "fall", 4 : "winter"})
hour_df['weathersit'] = hour_df['weathersit'].replace({1 : "Clear", 2 : "Mist + Cloudy", 3 : "Light Snow", 4 : "Heavy Rain + Ice Pallets"})
hour_df['holiday'] = hour_df['holiday'].replace({0 : "No", 1 : "Yes"})
hour_df['workingday'] = hour_df['workingday'].replace({0 : "No", 1 : "Yes"})
hour_df['weekday'] = hour_df['weekday'].replace({0 : "Sunday", 1 : "Monday", 2 : "Tuesday", 3 : "Wednesday", 4 : "Thursday", 5 : "Friday", 6 : "Saturday"})
hour_df['mnth'] = hour_df['mnth'].replace({1 : "January", 2 : "February", 3 : "March", 4 : "April", 5 : "May", 6 : "June", 7 : "July", 8 : "August", 9 : "September",
                                         10 : "October", 11 : "November", 12 : "December"})
```

```python
hour_df.head()
```

## Exploratory Data Analysis (EDA)

### Explore ...

#### Pertanyaan 1

```python
hour_per_casual = hour_df.groupby(by = "hr")['casual'].mean()
```

```python
hour_per_casual
```

#### Pertanyaan 2

```python
usage_by_holiday = day_df.groupby('holiday')['cnt'].mean()
```

```python
usage_by_holiday.head()
```

```python
day_df['holiday_encoded'] = day_df['holiday'].map({'No': 0, 'Yes': 1})

correlation = day_df['holiday_encoded'].corr(day_df['cnt'])
```

```python
print("Korelasi antara 'holiday' dan 'cnt':", correlation)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.bar(hour_per_casual.index, hour_per_casual.values)
plt.title('Jumlah Pengguna Casual per Jam')
plt.xlabel('Jam (hr)')
plt.ylabel('Jumlah Pengguna Casual')
plt.show()
```

### Pertanyaan 2:

```python
usage_by_holiday.plot(kind='bar', color=['red', 'green'])
plt.title('Pengaruh Hari Libur terhadap Jumlah Penggunaan Sepeda')
plt.xlabel('Hari Libur')
plt.ylabel('Rata-rata Jumlah Penggunaan Sepeda')
plt.show()
```
